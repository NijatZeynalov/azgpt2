```python
from bs4 import BeautifulSoup
import requests
from tqdm import tqdm
```

```python
category_name = 'dunya'
```

```python
# scrape categories

base_url = f'https://lent.az/{category_name}'
reqs = requests.get(base_url)
soup = BeautifulSoup(reqs.text, 'html.parser')
 
category_pages = []
for link in soup.find_all('a'):
    
    cat_links = link.get('href')
    if cat_links is not None and cat_links.startswith(f'{base_url}?page='):
        category_pages.append(cat_links)
        
last_page = max([int(i.split('=')[1]) for i in category_pages])
```

```python
category_pages
```

```python
news_pages = []
for item in range(1,last_page+1):
        news_pages.append(f'{base_url}?page={item}')
```

```python
news_pages
```

```python
news_links = []

for base_url in tqdm(news_pages):
    
    reqs = requests.get(base_url)
    soup = BeautifulSoup(reqs.text, 'html.parser')
    
    for link in soup.find_all('a'):

        news_link = link.get('href')

        if news_link is not None and news_link.startswith(f'https://lent.az/xeber/'):
            news_links.append(news_link)

news_links = list(set(news_links))
```

```python
len(news_links)
```

```python
import pandas as pd
pd.DataFrame(news_links).to_csv(f'{base_url}_urls.csv')
```
