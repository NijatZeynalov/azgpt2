```python
import requests
from bs4 import BeautifulSoup


links_and_page_numbers = {}

for url in urls:
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    pagination_class = soup.find('ul', class_='pagination')
    page_links = pagination_class.find_all('a')
    last_page_link = page_links[-2]  # The second to last link is the link to the last page
    last_page_number = int(last_page_link.get('href').split('=')[-1])

    links_and_page_numbers[url] = last_page_number

    print(f'Number of pages for {url}: {last_page_number}')

print(links_and_page_numbers)
```

```python
import requests
from bs4 import BeautifulSoup

urls = ['https://kulis.az/kulis+?page=', 'https://kulis.az/nesr?page=', 'https://kulis.az/senet?page=', 'https://kulis.az/edebi-tenqid?page=', 'https://kulis.az/media?page=']

page_links = []

for url in urls:
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    pagination_class = soup.find('ul', class_='pagination')
    page_links.extend([url + str(i) for i in range(1, int(pagination_class.find_all('a')[-2].get('href').split('=')[-1]) + 1)])

print(page_links)
```

```python
len(page_links)
```

```python
links = []
for url in tqdm(page_links):
    response = requests.get(url)

    soup = BeautifulSoup(response.text, 'html.parser')

    

    for link in soup.find_all('a'):
        news_link = link.get('href')
        if news_link.startswith('https://kulis.az/xeber'):
            links.append(news_link)
```

```python
len(set(links))
```

```python
news_links = list(set(links))
```
